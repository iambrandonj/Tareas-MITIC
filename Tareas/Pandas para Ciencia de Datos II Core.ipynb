### Transformación de Datos

```python
df['Total w/ Tax'] = df['Total'] + df['Tax 5%']
df[['Total', 'Tax 5%', 'Total w/ Tax']]
```

```python
grouped_product = df.groupby('Product line')
grouped_city = df.groupby('City')
grouped_customer = df.groupby('Customer type')
grouped_gender = df.groupby('Gender')
```

### Análisis de Datos
- Realiza agrupaciones de datos utilizando groupby para obtener insights específicos (por ejemplo, ventas por producto, ventas por región, etc.).
- Aplica funciones de agregación como sum, mean, count, min, max, std, y var.
-Utiliza el método apply para realizar operaciones más complejas y personalizadas.


```python
ventas_por_producto = grouped_product['Total'].sum()
print(ventas_por_producto)
```

```python
ventas_por_ciudad = grouped_city['Total'].sum()
print(ventas_por_ciudad)
```

```python
promedio_tax_genero = grouped_gender['Tax 5%'].mean()
print(promedio_tax_genero)
```

```python
cantidad_ventas_customer = grouped_customer['Total'].count()
print(cantidad_ventas_customer)
```

```python
df['Daytime'] = df['Time'].apply(lambda x: 'Day' if 6 <= x.hour < 18 else 'Night')
df[['Time', 'Daytime']]
```

```python
df['Rating Describe'] = df['Rating'].apply(lambda x: 'Good' if x >= 5 else 'Bad')
df[['Rating', 'Rating Describe']]
```

```python
df[df['Rating Describe'] == 'Bad']
```
